##### Q4. Write a program that prompts the user to enter a loan amount, annual interest rate, and number of years  for a car loan. Then calculate and print the monthly payment amount.

Note:
```
PMT= [PV*i*((1+i)^n)]/[((1+i)^n)-1]
```
PMT is the monthly payment

PV is the loan amount

i is the interest rate per month in decimal form (interest rate percentage divided by 12)

n is the number of months (term of the loan in months)

```
Loan Amount: 30000.00
Annual Interest Rate: 4.00
Years: 5
Your monthly payment is: $552.50
```

````
# Type solution here
loanAmount = int(input(print("Loan Amount: ")))
annualInterestRate = int(input(print("Annual Interest Rate: ")))
monthlyInterestRate = (annualInterestRate*0.01)/12
years = int(input(print("Years: ")))
months = years*12

PMT = (loanAmount*monthlyInterestRate*((1 + monthlyInterestRate)**months))/(((1+monthlyInterestRate)**months)-1)

pmtRound = round(PMT, 2)

print("Your monthly payment is:", "$" + str(pmtRound))
````

##### Q5. Write a program to that prompts the user to enter a string. Then check whether or not the string is a palindrome.

```
Enter a string: Madam
Madam, is a palindrome!
```

````
# Type solution here

string = input(print("Enter a string: "))
stringLower = string.lower()

if stringLower == stringLower[::-1]:
  print(string, "is a palindrome!")
else:
  print(string, "is not a palindrome!")

````

##### Q6. Write a program to compute and print all possible combinations of change for \$1. Denominations to be considered – quarter, dime, nickel, penny.

Hint: Use nested loops (loops within loops for the various denominations of coins)

```
Change for $1:
0 quarters, 0 dimes, 0 nickels, 100 pennies
…
4 quarters, 0 dimes, 0 nickels, 0 pennies
```

````
# Type solution here
print("Change for a $1: ")

for quarter in range(5):
  for dime in range(11):
    for nickel in range(21):
      for penny in range(101):
        if ((25*quarter)+(10*dime)+(5*nickel)+(1*penny))==100:
          print(quarter, "quarters,", dime, "dimes,", nickel, "nickels,", penny, "pennies")








````

##### Q8. Write a program to ask the user to enter a password. Then check to see if it is a valid password based on these requirements -

Must be at least 8 characters long

Must contain both uppercase and lowercase letters

Must contain at least one number between 0-9

Must contain a special character -!,@,#,$

If the password is not valid, ask the user to re-enter. This should continue until the user enters a valid password. After a valid password is entered, print Access Granted!

```
Please enter a password. Follow these requirements - 
a. Must be at least 8 characters long
b. Must contain both uppercase and lowercase letters
c. Must contain at least one number between 0-9
d. Must contain a special character -!,@,#,$
Password: HelloWorld1
Invalid password. Try again!
Password: Hello@World1
Access Granted!
```


````
# Type solution here

print("Please enter a password. Follow these requirements - ")
print("a. Must be atleast 8 characters long")
print("b. Must contain both uppercase and lowercase letters")
print("c. Must contain at least one number between 0-9")
print("d. Must contain a special character - !, @, #, $")

passwordAccess = False
while passwordAccess == False:
  password = input(print("Password: "))

  lower = "abcdefghijklmnopqrstuvwxyz"
  digits = "0123456789"
  upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  specialCharacter = "!@#$"

  yesSpecialCharacter = False
  minLenght = False
  lowerCase = False
  upperCase = False
  yesDigit = False
  
  if len(password) >= 8:
    minLenght = True

  for x in range(len(password)):
    if password[x] in upper:
      upperCase = True
    if password[x] in lower:
      lowerCase = True
    if password[x] in digits:
      yesDigit = True
    if password[x] in specialCharacter:
      yesSpecialCharacter = True

  if minLenght == True and lowerCase == True and upperCase == True and yesDigit == True and yesSpecialCharacter == True:
    print("Access Granted!")
    passwordAccess = True
  else:
    print("Invalid password, try again!")
    passwordAccess = False
    
    
````

##### Q9. Given the score sheet of participants in a competition, you are required to find the runner-up score. You are given scores in a list. Find the score of the runner-up.

```
list1 - [2, 3, 6, 6, 5]
runner_up - 5

list2 - [1, 7, 5, 3, 10, 4, 5, 5, 6, 11]
runner_up - 10
```

````
# Type solution here

list = [2, 3, 6, 6, 5]
print("list1 - ", list)
list.sort()
maxInList = max(list)
while max(list) == maxInList:
  list.remove(max(list))
print("runner up - ", max(list))

list2 = [1, 7, 5, 3, 10, 4, 5, 5, 6, 11]
print("list2 - ", list2)
list2.sort()
maxInList2 = max(list2)
while max(list2) == maxInList2:
  list2.remove(max(list2))
print("runner up - ", max(list2))
````
